```python
import numpy as np
```

```python
# buat vector x dengan ukuran 3x1
x = np.array([54, 5, 4])
print("Vektor x :", x)
```

```python
# hitung nilai L2 norm dari vector x
print("Nilai L2 norm vektor x :", np. linalg.norm(x))
```

```python
# buat vector v dengan elemen (3,4)
v = np.array([3, 4])
print("Vektor v :", v)
```

```python
# hitung nilai L2 norm dari vector v
print("Nilai L2 norm vektor v :", np.linalg.norm(v))
```

```python
# tampilkan unit vector dari vector v
v = np.array([3, 4])
unit_v = v / np.linalg.norm(v)
print("Unit Vektor v :", unit_v)
```

```python
# hitung panjang unit vector dari vector v
v = np.array([3, 4])
unit_v = v / np.linalg.norm(v)
print("Panjang unit vector v :", np.linalg.norm(unit_v))
```
