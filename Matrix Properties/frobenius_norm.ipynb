# Using numpy

```python
import numpy as np
```

```python
X = np.array([[6, 2, 1], [2, 5, 1], [6, 6, 1]])
X
```

```python
X = np.array([[1, 2], [3, 4]])
X
```

```python
(6**2 + 2**2 + 6**2 + 2**2 + 5**2 + 6**2 + 1**2 + 1**2 + 1**2)**(1/2)
```

```python
np.linalg.norm(X) # same function as for vector L2 norm
```

# Using pytoch

```python
import torch
```

```python
X_pt = torch.tensor([[6, 2, 1], [2, 5, 1], [6, 6, 1]]) # torch.norm() supports floats only
X_pt
```

```python
torch.norm(X_pt.float())
```

# Using TensorFlow

```python
import tensorflow as tf
```

```python
X_tf = tf.Variable([[6, 2, 1], [2, 5, 1], [6, 6, 1]], dtype=tf.float32)
X_tf
```

```python
tf.norm(X_tf)
```
