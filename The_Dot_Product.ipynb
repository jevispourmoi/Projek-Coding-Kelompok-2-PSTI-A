```python
#buat vector x dan y dengan ukuran 1x3 menggunakan numpy
import numpy as np
```

```python
X = np.array([25, 15, 61])
print("X=", X)
```

```python
Y = np.array([12, 4, 7])
print("Y=", Y)
```

```python
#hitung perkalian dot product antara vector x dan vector y
np.dot(X, Y)
print("X.Y=", np.dot(X, Y))
```

```python
#buat vector x_pt dan y_pt dengan ukuran 1x3 menggunakan pytorch
import torch
```

```python
X_pt = torch.tensor([25, 15, 61])
print("X=", X_pt)
```

```python
Y_pt = torch.tensor([12, 4, 7])
print("Y=", Y_pt)
```

```python
#hitung perkalian dot product antara vector x_pt dan vector y_pt
torch.dot(X_pt, Y_pt)
print("X.Y=",torch.dot(X_pt, Y_pt))
```

```python
#buat vector x_tf dan y_tf dengan ukuran 1x3 menggunakan tensorflow
import tensorflow as tf
```

```python
X_tf = tf.Variable([25, 15, 61])
print("X=", X_tf)
```

```python
Y_tf = tf.Variable([12, 4, 7])
print("Y=", Y_pt)
```

```python
tf.reduce_sum(tf.multiply(X_tf, Y_tf))
print("X.Y=", tf.reduce_sum(tf.multiply(X_tf, Y_tf)))
```

```python
# 2. what is the hadamard product of these matrices?
A = torch.tensor([[25, 10], [-2, 1]])
print("A =", A)
```

```python
B = torch.tensor([[-1, 7], [10, 8]])
print("B =", B)
```

```python
torch.sum(torch.mul(A, B))
print("hadamard product dari matrix A dan B adalah:", torch.sum(torch.mul(A, B)))
```
